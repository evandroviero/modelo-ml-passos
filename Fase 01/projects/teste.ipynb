```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Exemplo de URL (substitua pela real se tiver uma)
ano = 2023
url = f'http://vitibrasil.cnpuv.embrapa.br/index.php?ano={ano}&opcao=opt_02'

# Se for local, use diretamente a string HTML que você colou
# html = '''...'''  # Coloque seu HTML aqui
# soup = BeautifulSoup(html, 'html.parser')

# Se for da web:
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Encontra a tabela com a classe especificada
tabela = soup.find('table', class_='tb_base tb_dados')

# Inicializa listas para armazenar os dados
dados = []
categoria_atual = None

# Itera pelas linhas do tbody
for tr in tabela.tbody.find_all('tr'):
    cols = tr.find_all('td')
    if not cols:
        continue

    nome = cols[0].get_text(strip=True)
    quantidade = cols[1].get_text(strip=True)

    if 'tb_item' in cols[0].get('class', []):
        # Nova categoria principal
        categoria_atual = nome
        dados.append({'Categoria': categoria_atual, 'Produto': '', 'Quantidade (L)': quantidade})
    elif 'tb_subitem' in cols[0].get('class', []):
        # Produto dentro da categoria atual
        dados.append({'Categoria': categoria_atual, 'Produto': nome, 'Quantidade (L)': quantidade})

# Cria DataFrame
df = pd.DataFrame(dados)

# Exibe os dados
# df

# Se quiser salvar em CSV
# df.to_csv('vinhos_producao.csv', index=False)
```
